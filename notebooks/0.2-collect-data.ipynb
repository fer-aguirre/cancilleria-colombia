```python
import requests
import pandas as pd
from bs4 import BeautifulSoup

df = pd.read_csv('../data/processed/cancilleria_extendida.csv')

df_filtered = df[df['Url'].notna()]

df_filtered.shape
```

```python

nombres = df_filtered['Nombre'].to_list()

urls = df_filtered['Url'].to_list()

dfs = []

for i in range(len(urls)):

    response = requests.get(urls[i])

    html = response.text

    # Parse html
    soup = BeautifulSoup(html, 'html.parser')

    table = soup.find('table', attrs={'class': 'directorio_tabla'})

    df_table = pd.read_html(str(table), header=0)[0]

    df_table['Nombre'] = nombres[i]

    dfs.append(df_table)

df_join = pd.concat(dfs)

df_join
```

```python
df_join.to_csv('../data/processed/funcion_publica_extendida.csv', index=None)
```
