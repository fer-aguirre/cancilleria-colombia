```python
# Importing the required modules
import pandas as pd
import numpy as np
from bs4 import BeautifulSoup
import glob

# Create a list with all the files in the directory with 'html' extension
files = glob.glob('*.html')

data = []

# Iterate trough the elements on the list
for i in range(len(files)):

    # Parse 'html' content with BeautifoulSoup
    soup = BeautifulSoup(open(files[i]),'html.parser')

    # Get data from 'td' tags
    table = soup.find_all("td", attrs={'class':'cajax'})

    # Create a list with data extracted
    data_raw = [x.text for x in table]

    # Clean data
    data_clean = [np.nan if x == '' else x for x in data_raw]

    # Group data by 10 elements
    data_proccessed = [data_clean[n:n+10] for n in range(0, len(data_clean), 10)]

    # Append list to list
    data.append(data_proccessed)

# Flatten lists of lists
flat_list = [item for sublist in data for item in sublist]

# Convert list to dataframe and assign column names
df = pd.DataFrame(flat_list, columns = ['Nombre', 'Cargo', 'Dependencia', 'Continente', 'País', 'Ciudad', 'Correo', 'Número', 'CP', 'Grupo_Trabajo'])

# Save dataframe as 'csv' file
df.to_csv('../data/raw/cancillerias.csv', index=False)
df
```
